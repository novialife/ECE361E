```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
vgg11 = open('vgg11.txt', 'r')
vgg11 = vgg11.read()
vgg11 = vgg11.split('\n')
vgg11_train_loss = vgg11[34]
vgg11_train_acc = vgg11[35]
vgg11_test_loss = vgg11[36]
vgg11_test_acc = vgg11[37]
```

```python
vgg16 = open('vgg16.txt', 'r')
vgg16 = vgg16.read()
vgg16 = vgg16.split('\n')
vgg16_train_loss = vgg16[46]
vgg16_train_acc = vgg16[47]
vgg16_test_loss = vgg16[48]
vgg16_test_acc = vgg16[49]
```

```python
mbn = open('mbn.txt', 'r')
mbn = mbn.read()
mbn = mbn.split('\n')
mbn_train_loss = mbn[87]
mbn_train_acc = mbn[88]
mbn_test_loss = mbn[89]
mbn_test_acc = mbn[90]
```

```python
vgg11_test_acc = vgg11_test_acc.split(' ')[2:]
vgg11_test_acc = pd.Series(vgg11_test_acc)
vgg11_test_acc = vgg11_test_acc.str[:-1]
vgg11_test_acc[0] = vgg11_test_acc[0][1:]
vgg11_test_acc = vgg11_test_acc.astype(float)
```

```python
vgg16_test_acc = vgg16_test_acc.split(' ')[2:]
vgg16_test_acc = pd.Series(vgg16_test_acc)
vgg16_test_acc = vgg16_test_acc.str[:-1]
vgg16_test_acc[0] = vgg16_test_acc[0][1:]
vgg16_test_acc = vgg16_test_acc.astype(float)
```

```python
mbn_test_acc = mbn_test_acc.split(' ')[2:]
mbn_test_acc = pd.Series(mbn_test_acc)
mbn_test_acc = mbn_test_acc.str[:-1]
mbn_test_acc[0] = mbn_test_acc[0][1:]
mbn_test_acc = mbn_test_acc.astype(float)
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(vgg11_test_acc, label='VGG11')
plt.plot(vgg16_test_acc, label='VGG16')
plt.plot(mbn_test_acc, label='MobileNetV1')
plt.xlabel('Epoch')
plt.ylabel('Test Accuracy (%)')
plt.title('Test Accuracy of VGG11 and VGG16')
plt.legend()
plt.show()
```
